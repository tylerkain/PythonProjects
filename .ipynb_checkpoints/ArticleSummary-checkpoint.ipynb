```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
article_url = "https://www.washingtonpost.com/politics/republicans-midterms-digital-influence/2021/04/03/391103a8-8c0e-11eb-a730-1b4ed9656258_story.html"
```

```python
page = urlopen(article_url).read().decode('utf8')
soup = BeautifulSoup(page, features='lxml')
soup
```

```python
soup.find('article')
```

```python
text = ' '.join(map(lambda p: p.text, soup.find_all()))
```

```python
def GetText(url):
    page = urlopen(url).read().decode('utf8')
    soup = BeautifulSoup(page, features='lxml')
    text = " ".join(map(lambda p: p.text, soup.find_all()))
    return text.replace("\n", " ")
```

```python
text = GetText("https://www.washingtonpost.com/politics/republicans-midterms-digital-influence/2021/04/03/391103a8-8c0e-11eb-a730-1b4ed9656258_story.html")
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
from string import punctuation
```

```python
sents = sent_tokenize(text)
sents
```

```python
word_sent = word_tokenize(text)
word_sent
```

```python
_stopwords = set(stopwords.words('english') + list(punctuation))
_stopwords
```

```python
word_sent =[word for word in word_sent if word not in _stopwords]
```

```python
word_sent
```

```python
from nltk.probability import FreqDist
freq = FreqDist(word_sent)

freq
```

```python
from heapq import nlargest
```

```python
nlargest(10, freq, key=freq.get)
```

```python
from collections import defaultdict
```

```python
ranking = defaultdict(int)

for i, sent in enumerate(sents):
    for w in word_tokenize(sent.lower()):
        if w in freq:
            ranking[i] += freq[w]
```

```python
defaultdict(int)
```

```python
ranking
```

```python
sents_idx = nlargest(4, ranking, key=ranking.get)
sents_idx
```

```python
[sents[j] for j in sorted(sents_idx)]
```
